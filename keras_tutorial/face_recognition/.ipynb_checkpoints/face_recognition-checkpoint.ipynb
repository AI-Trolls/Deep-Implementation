# 케라스(keras)로 나만의 데이터셋을 이용하여 얼굴 인식 학습하기

```python
from keras import layers
from keras import models

model = models.Sequential()
model.add(layers.Conv2D(32, (5, 5), strides=1, input_shape=(32, 32, 3)))
model.add(layers.Conv2D(32, (5, 5), activation='relu', strides=2, input_shape=(96, 96, 3)))
model.add(layers.Conv2D(32, (5, 5), strides=1, input_shape=(64, 64, 3)))
model.add(layers.Conv2D(32, (5, 5), activation='relu', strides=2, input_shape=(96, 96, 3)))
model.add(layers.Conv2D(32, (5, 5), strides=1, input_shape=(128, 128, 3)))
model.add(layers.Conv2D(32, (5, 5), activation='relu', strides=2, input_shape=(96, 96, 3)))
model.summary()
```
